```python
import re
import string

def deep_cleaner(source):
        
        source = source.replace(' da ',' ')
        source = source.replace(' da,',', ')
        source = source.replace(" de " , " ")
        source = source.replace(" vs " , " ")
        source = source.replace(" ve " , " ")
        return source
```

```python
#removing punctuations
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
def remove_dig(text):
    for digits in string.digits:
        text = text.replace(digits,'')
    return text
```

```python
from statistics import mean
import numpy as np

xs = np.array(df['retweet_count'], dtype=np.float64)
ys = np.array(df['favorite_count'], dtype=np.float64)

def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    
    b = mean(ys) - m*mean(xs)
    
    return m, b

m, b = best_fit_slope_and_intercept(xs,ys)

print(m,b)
```

```python
regression_line = [(m*x)+b for x in xs]
```

```python
regression_line = []
for x in xs:
    regression_line.append((m*x)+b)
```
