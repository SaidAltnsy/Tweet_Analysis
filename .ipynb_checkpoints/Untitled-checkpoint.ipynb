```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(x,y,label = 'asd')
axes.plot(x,x**3, label = 'trd')
axes.legend(loc =(0.1,0.1))
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)
axe1 = fig.add_axes([0.5,0.8,0.3,0.3])
axe1.plot(x,x**4)
```
